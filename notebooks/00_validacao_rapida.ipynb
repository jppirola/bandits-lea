# validação rápida

este notebook serve para verificar se o pacote está instalado, se as políticas executam sem erro e se os resultados têm o formato esperado.

usar poucas replicações aqui. deixar a simulação completa para o notebook seguinte ou para o script `scripts/run_simulation.py`.

```python
from pathlib import Path
import sys

# adicionar src ao caminho quando o notebook for executado sem instalar o pacote
PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(PROJECT_ROOT / "src"))
```

```python
from bandit_inference.analysis import summarize_by_policy
from bandit_inference.simulation import ExperimentConfig, run_many_experiments
```

## configurar uma simulação pequena

usar horizonte e número de replicações pequenos apenas para validar a execução.

```python
# definir configuração pequena para teste
config = ExperimentConfig(
    n_arms=5,
    horizon=100,
    n_replications=50,
    sigma=1.0,
    seed=12345,
)

# executar simulação rápida
results = run_many_experiments(config=config, progress=True)
results.head()
```

```python
# resumir resultados por política
summary = summarize_by_policy(results)
summary
```

## checar consistência básica

verificar se existem quatro políticas, se a cobertura está entre 0 e 1 e se todos os braços receberam pelo menos uma observação.

```python
# validar número de políticas
assert results["policy"].nunique() == 4

# validar limites da cobertura
assert summary["empirical_coverage"].between(0, 1).all()

# validar presença de observações em todos os braços
count_columns = [column for column in results.columns if column.startswith("n_arm_")]
assert (results[count_columns] > 0).all().all()

print("validação concluída")
```
